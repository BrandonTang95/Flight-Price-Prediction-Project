```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
import warnings
import os
```

```python
# Suppress warnings for cleaner output
warnings.filterwarnings('ignore')

# Set Pandas display options to see all columns clearly in console
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)
```

```python
df = pd.read_csv("flight-price-prediction/clean_data_v2.csv")
df
```

```python
sns.barplot(data=df, x='Airline', y='Price', palette='viridis')
plt.xlabel('Airline')
plt.ylabel('Price (in INR)')
plt.xticks(rotation=45)
plt.title('Airline vs Price')
plt.show()
```

```python
ax = sns.countplot(data=df, x='Stops', palette='mako')
for container in ax.containers:
    ax.bar_label(container)
    
plt.title('Number of Stops Distribution')
plt.xlabel('Number of Stops')
plt.ylabel('Count')
plt.show()
```
